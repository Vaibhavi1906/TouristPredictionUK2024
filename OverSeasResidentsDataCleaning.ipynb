```python
merged2_data = merged_data
def backward_fill_values(data, year_to_fill, quarters, column_to_fill):
    """
    Fills missing values in the specified year and quarters using backward fill (bfill) method,
    looping through each Area of Residence and Purpose of Visit.

    Parameters:
    data (pd.DataFrame): The input dataframe containing the data.
    year_to_fill (str): The year for which to fill missing values (e.g., '2020').
    quarters (list): List of quarters to fill (e.g., ['Q2', 'Q3', 'Q4']).
    column_to_fill (str): The column name where missing values should be filled.

    Returns:
    pd.DataFrame: DataFrame with missing values filled using backward fill.
    """

    # Loop through each area of residence
    for area in data['Area of Residence'].unique():
        # Loop through each purpose of visit within the area of residence
        for purpose in data[data['Area of Residence'] == area]['Purpose of Visit'].unique():
            print(f"Processing Area: {area}, Purpose: {purpose}")
            
            for quarter in quarters:
                # Create the period to fill
                period_to_fill = f'{year_to_fill} {quarter}'
                
                # Check if the value is missing
                # is_missing = data.loc[
                #     (data['Period'] == period_to_fill) & 
                #     (data['Area of Residence'] == area) & 
                #     (data['Purpose of Visit'] == purpose), 
                #     column_to_fill
                # ].isna().any()
                
                #
                    # Find the next period (next year, same quarter)
                next_year = str(int(year_to_fill) + 1)
                next_period = f'{next_year} {quarter}'
                    
                    # Get the value from the next period
                next_value = data.loc[
                        (data['Period'] == next_period) & 
                        (data['Area of Residence'] == area) & 
                        (data['Purpose of Visit'] == purpose), 
                        column_to_fill
                    ].values
                    
                    # Ensure there's a value to backfill from the next period
                if len(next_value) > 0 and not pd.isna(next_value[0]):
                        data.loc[
                            (data['Period'] == period_to_fill) & 
                            (data['Area of Residence'] == area) & 
                            (data['Purpose of Visit'] == purpose), 
                            column_to_fill
                        ] = next_value[0]
                        
                        print(f"Filled {period_to_fill} with {next_period} value: {next_value[0]}")

    return data

# Apply the backward fill function to each relevant column for 2020 Q2, Q3, Q4
quarters_to_impute = ['Q1','Q2','Q3','Q4']
columns_to_impute = ['Total Visits', 'Total Spending in millions', 'Nights Spent','Visits by Purpose','Spending in millions','Number of visits by Air','Number of visits by Sea and Tunnel']

for quarter in quarters_to_impute:
    for column in columns_to_impute:
        merged_data = backward_fill_values(merged_data, '2020', [quarter], column)

# Save the updated data to a new CSV file
output_path = 'C://Users/vaibh/OneDrive/Desktop/BUSINESSPROJECT/Backfilled_TravelTrends.csv'
merged_data.to_csv(output_path, index=False)

```

```python
def aggregate_quarterly_to_yearly(data, year, columns_to_aggregate):
    """
    Aggregates quarterly data into yearly data by summing up values across all quarters for a given year.

    Parameters:
    data (pd.DataFrame): The input dataframe containing the data.
    year (str): The year for which to aggregate data (e.g., '2020').
    columns_to_aggregate (list): List of column names to aggregate.

    Returns:
    pd.DataFrame: DataFrame with yearly aggregated data.
    """
    # Initialize a list to collect aggregated rows
    yearly_data_rows = []

    # Loop through each area of residence
    for area in data['Area of Residence'].unique():
        # Loop through each purpose of visit within the area of residence
        for purpose in data[data['Area of Residence'] == area]['Purpose of Visit'].unique():
            # Filter the data for the specific area, purpose, and year
            filtered_data = data[
                (data['Area of Residence'] == area) &
                (data['Purpose of Visit'] == purpose) &
                (data['Period'].str.startswith(year) & data['Period'].str.contains('Q'))
            ]

            # Aggregate the data by summing up the quarterly values
            aggregated_row = {
                'Period': year,
                'Area of Residence': area,
                'Purpose of Visit': purpose
            }

            for column in columns_to_aggregate:
                # Sum the values across the quarters for this specific area, purpose, and year
                aggregated_row[column] = filtered_data[column].sum()

            # Append the aggregated row to the list
            yearly_data_rows.append(aggregated_row)
    
    # Convert the list of rows into a DataFrame
    yearly_data = pd.DataFrame(yearly_data_rows)
    
    return yearly_data

# Columns to aggregate
columns_to_aggregate = ['Total Visits', 'Total Spending in millions', 'Nights Spent', 'Visits by Purpose', 'Spending in millions','Number of visits by Air','Number of visits by Sea and Tunnel']

# Aggregate the data for the year 2020
yearly_aggregated_data = aggregate_quarterly_to_yearly(merged_data, '2020', columns_to_aggregate)

# Replace the existing empty yearly rows in the original DataFrame with the aggregated data
for index, row in yearly_aggregated_data.iterrows():
    merged_data.loc[
        (merged_data['Period'] == row['Period']) &
        (merged_data['Area of Residence'] == row['Area of Residence']) &
        (merged_data['Purpose of Visit'] == row['Purpose of Visit']),
        columns_to_aggregate
    ] = row[columns_to_aggregate].values

# Save the updated data to a new CSV file
output_path = 'OverseasResidents_CleanedData.csv'
merged_data.to_csv(output_path, index=False)
```
