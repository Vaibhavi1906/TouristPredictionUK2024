```python
import pandas as pd

# Load each relevant sheet
sheet9_df = pd.read_excel('TravelTrendsToUK.xlsx', sheet_name='9', skiprows=2)
sheet10_df = pd.read_excel('TravelTrendsToUK.xlsx', sheet_name='10', skiprows=2)
sheet11_df = pd.read_excel('TravelTrendsToUK.xlsx', sheet_name='11', skiprows=2)
```

```python
print("Column names in sheet 11:", sheet11_df.columns.tolist())

try:
    NightsSpentByRegion_df = sheet11_df.melt(id_vars=['Region of UK', 'Area of Residence'], var_name='Year', value_name='Nights Spent by region of UK and area of residence')
    print(NightsSpentByRegion_df.tail())
    print("Data extraction successful.")
except KeyError as e:
    print(f"Column error: {e}")
except Exception as e:
    print(f"Unexpected error: {e}")
```

```python
# Ensure the 'Year' column is treated as a string
Region_df['Year'] = Region_df['Year'].astype(str)

# Replace "Not available" with NaN
Region_df['Nights Spent by region of UK and area of residence'] = Region_df['Nights Spent by region of UK and area of residence'].replace("Not Available", np.nan)
Region_df['Visits by region of UK and area of residence'] = Region_df['Visits by region of UK and area of residence'].replace("Not available", np.nan)
Region_df['Spendings by region of UK and area of residence'] = Region_df['Spendings by region of UK and area of residence'].replace("Not available", np.nan)

# Print the DataFrame before backfill for 2020
print("Before Backfill - Data for 2020:")
print(Region_df[Region_df['Year'] == '2020'])

# Proceed with the backfill if data for 2020 is present
def backward_fill_values(data, year_to_fill, quarters, column_to_fill):
  # Loop through each region
    for region in data['Region of UK'].unique():
        # Loop through each area of residence within the region
        for area in data[data['Region of UK'] == region]['Area of Residence'].unique():
            print(f"Processing Region: {region}, Area of Residence: {area}")

            for quarter in quarters:
                # Create the period to fill
                period_to_fill = f'{year_to_fill} {quarter}'
                next_year = str(int(year_to_fill) + 1)
                next_period = f'{next_year} {quarter}'

                # Get the value from the next period
                next_value = data.loc[
                    (data['Year'] == next_year) & 
                    (data['Region of UK'] == region) & 
                    (data['Area of Residence'] == area), 
                    column_to_fill
                ].values

                # Ensure there's a value to backfill from the next period
                if len(next_value) > 0 and not pd.isna(next_value[0]):
                    data.loc[
                        (data['Year'] == year_to_fill) & 
                        (data['Region of UK'] == region) & 
                        (data['Area of Residence'] == area), 
                        column_to_fill
                    ] = next_value[0]

                    print(f"Filled {period_to_fill} with {next_period} value: {next_value[0]}")

    return data

# Define the quarters and columns to fill
quarters_to_impute = ['Q1', 'Q2', 'Q3', 'Q4']
columns_to_impute = [
    'Visits by region of UK and area of residence',
    'Spendings by region of UK and area of residence',
    'Nights Spent by region of UK and area of residence'
]

# Apply the backward fill function for 2020
for column in columns_to_impute:
    Region_df = backward_fill_values(Region_df, '2020', quarters_to_impute, column)

# Print the entire DataFrame after backfill to see all data for 2020
print("After Backfill - Entire DataFrame for 2020:")
print(Region_df[Region_df['Year'] == '2020'])
    # Save the updated data to a new CSV file
output_path = 'RegionBased_TravelTrends.csv'
Region_df.to_csv(output_path, index=False)
```
